```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
```

```python
titanic.info()
```

1.Create a pie chart presenting the male/female proportion

```python
persons = titanic.groupby('sex').size()
print (persons)
plt.axis('equal')
plt.pie(persons, labels=persons.index)
```

2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
females = titanic.loc[titanic['sex'] == 'female']
males = titanic.loc[titanic['sex'] == 'male']
plt.scatter(females['age'],females['fare'])
plt.scatter(males['age'],males['fare'])
```
